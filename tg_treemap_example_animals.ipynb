```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
images = os.listdir('./images')
layer3['image'] =  layer3['idx']+'-' + layer3['h3'].str.lower() + 'x.gif'
layer3['image'] = np.where(layer3['image'].isin(images),layer3['image'],None)

```

```python
fig = p.draw()
ax = fig.axes[0]

l3 = layer3.copy().dropna(subset='image')
for i, row in l3.iterrows():
    img = matplotlib.offsetbox.OffsetImage(
        plt.imread('./images/' + row['image']), 
        zoom = .5)
    ax.add_artist(matplotlib.offsetbox.AnnotationBbox(img,
        (row['x']+row['dx']/2 ,row['y']+row['dy']/2-30)
        ,pad = 0.0
        ,frameon =False
        ,zorder=20
        )
        )
fig
```
